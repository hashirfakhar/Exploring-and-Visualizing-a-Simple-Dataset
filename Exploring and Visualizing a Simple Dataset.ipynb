# Introduction

In this task, we explore the famous Iris dataset using pandas, seaborn, and matplotlib.
We aim to understand the structure of the data and visualize relationships between variables.

# Load Libraries & Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset
df = sns.load_dataset('iris')
```

# Dataset Overview

```python
print("Shape of dataset:", df.shape)
print("Columns:", df.columns.tolist())
df.head()
```

# Data Cleaning & Preparation

```python
# Check for missing values
print("\nMissing values:\n", df.isnull().sum())

# Check for duplicates
print("\nDuplicate rows:", df.duplicated().sum())

# Remove duplicates
df.drop_duplicates
```

# Exploratory Data Analysis (EDA)

## Histograms

```python
df.hist(figsize=(10, 6), edgecolor='black')
plt.suptitle('Feature Distributions', fontsize=16)
plt.tight_layout()
plt.show()
```

## Box Plots

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Boxplot for all numeric features')
plt.show()
```

## Scatter Plot (petal_length vs petal_width)

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Petal Width by Species')
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

# Key Observations

- Setosa species has clearly smaller petal length and width compared to the other two.
- Virginica tends to have larger petals.
- Sepal measurements are less useful for separation than petal measurements.
- Few minor outliers exist in petal width.
- High correlation between petal_length and petal_width (~0.96).

# Conclusion

This analysis gave us a visual and statistical understanding of the Iris dataset.
Petal features appear to be the most significant in distinguishing between species.
This forms a good foundation for building classification models in future tasks.
